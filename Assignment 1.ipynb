```python
import pandas as pd
import sys
import numpy as np
```

```python
#Load the data and programmatically define the column names
```

```python
adults=pd.read_csv('adult.csv',sep=' *, *',engine='python',names=["age","workclass","fnlwgt","education","education-num","marital-status","occupation","relationship","race","sex","capital-gain","capital-loss","hours-per-week","native-country","salary"])
```

```python
adults
```

```python
#Create an output containing mean, minimum and maximum hours per week for every country.
```

```python
country_hours=adults[['native-country','hours-per-week']]
Mean=country_hours.groupby('native-country').mean()
Min=country_hours.groupby('native-country').min()
Max=country_hours.groupby('native-country').max()
country_hours_comb=pd.concat([Mean,Min,Max],axis=1)
country_hours_comb.columns=['Mean','Min','Max']
country_hours_comb
```
